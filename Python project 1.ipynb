EXPLORATORY DATA ANALYSIS

```python
colors = ['#30CF17', '#D7EF0B']
# define the colors for the bars first
ax=sns.countplot(x ='Gender',data = df, palette = colors, hue= 'Gender',dodge = False)
# define x axis, datasource as df, palette is colors as defined above, put the x axis(columm) as the hue,put dodge as false which ensures that bar is not split when hue is used

for bars in ax.containers:
    ax.bar_label(bars)
# add the labels to the bars i.e the count

9# plot count bar chart for gender and its count 
```

```python
# plotting a bar chart for gender vs total amount

sales_gen= df.groupby(['Gender'], as_index= False)['Amount'].sum().sort_values(by = 'Amount', ascending =False)
#groups the data by gender and finds the sum of the amount for each gender group and  as_index=False parameter keeps the Gender column as a regular column rather than an index.
#ascending false asks to sort by descending order
#store this calculate in the sales_gen variable
colors = ['#E228EB', '#5716B0']
# define the colors needed for the bars
sns.barplot(x='Gender', y='Amount', data=sales_gen, hue='Gender', palette=colors, dodge=False)
# define x axis column and y axis column, datasource as sales_gen, palette is colors as defined above, put the x axis(columm) as the hue,put dodge as false which ensures that bar is not split when hue is used
plt.legend([], [], frameon=False)
#used to remove the legend from the plot.
```

```python
#bifurcation of age groups based on gender and their count plotted.
colors = ['#0D2EC2', '#0ADED4']
ax = sns.countplot(data = df, x = 'Age Group' , hue = 'Gender',palette = colors)
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_age = df.groupby(['Age Group'], as_index = False)['Amount'].sum().sort_values(by = 'Amount' ,ascending = False)
colors = ['#FF96D7','#D905C6','#730469','#72546F','#F3CCEF','#A07F9D','#4B2948']
rt=sns.barplot( x = 'Age Group', y = 'Amount',data = sales_age, hue = 'Age Group', palette = colors,dodge = False)
plt.legend([],[],frameon= False)
```

```python
# total sum of amount  from top 10 states
sales_state= df.groupby(['State'], as_index = False) ['Amount'].sum().sort_values( by = 'Amount', ascending = False).head(10)
sns.set(rc={'figure.figsize':(15,5)})
#This sets the default size of the plots created with seaborn to be 15 inches wide and 5 inches tall. 
#The rc parameter is a dictionary where 'figure.figsize' is a key that specifies the dimensions of the plot.
colors= ['#B24A26','#D65E35','#FA7C50','#F99B7B','#F8B198','#EDF540','#F4FA82','#F9FCAF','#E1E4A2','#FDFFD3']
rt =sns.barplot( x = 'Amount', y ='State', data=sales_state, hue = 'State',palette=colors, dodge=False)
plt.legend([],[],frameon=False)
for bars in rt.containers:
     rt.bar_label(bars)
```

```python
#total orders from the top 10 states
orders_state= df.groupby(['State'],as_index = False) ['Orders'].sum().sort_values(by = 'Orders', ascending = False).head(10)
sns.set(rc={'figure.figsize':(15,5)})
colors=['#B24A26','#D65E35','#FA7C50','#F99B7B','#F8B198','#EDF540','#F4FA82','#F9FCAF','#E1E4A2','#FDFFD3']
rt = sns.barplot(x ='Orders', y = 'State', data= orders_state, hue='State', palette=colors, dodge = False)
plt.legend([],[],frameon=False)
for bars in rt.containers:
     rt.bar_label(bars)
```
